##  Aufgabe 2: C^-1 = C^T

Definieren Sie für beliebiges N aus den natürlichen Zahlen die Matrix C und zeigen Sie, dass die Inverse von C gleich der Transponierten ist, d.h.
C^-1 = C^T
Dazu müssen Sie zeigen, dass gilt:
CC^T = C^T * C = 1

Kontrollieren Sie für N = 2; 4; 8 und N = 16!

```python
def proof_characteristic(N):
    C = np.zeros(shape=(N,N))
    
    for i in range(N):
        for j in range(N):
            if i == 0:
                c_n = np.sqrt(1/N)
            else:
                c_n = np.sqrt(2/N)
            C[i][j] = c_n * np.cos((2*j+1)*i*np.pi/(2*N))
    print('N: ', N)
    print(C)
    result = np.max(abs(LA.inv(C)-C.T))
    print('Inverse - Transponierte: ', result, '\n')

proof_characteristic(2)
proof_characteristic(4)
proof_characteristic(8)
proof_characteristic(16)
```
